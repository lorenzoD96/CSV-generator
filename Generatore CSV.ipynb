```python
def generate_week(daylist,goal_sum,size=7):
    output = []
    for i in range(size):
        day_noise = np.random.normal(0,1,100)
        base = daylist[i]*goal_sum
        bias = np.mean(day_noise)*base
        new_int = int(base+bias)
        output.append(new_int)
    return output
```

```python
generate_week(week_distrib_alt,200,7)
```

```python
def appendi_settimana(df):
    week = generate_week(week_distrib,200,7)
    for i in range(0,len(week)):
        df['GIORNO_'+str(i+1)] = week[i]
    return df    

def appendi_settimana_alt(df):
    week = generate_week(week_distrib_alt,200,7)
    for i in range(0,len(week)):
        df['GIORNO_'+str(i+1)] = week[i]
    return df    
```

```python
def genera_orari(df,day_distrib,qta_max,lista_farm):
    lista = []
    for i in range(0,7):        
        giorno = datetime.strptime(str(giorni[i]), '%Y-%m-%d')
        n_trans = df['GIORNO_'+str(i+1)]
        for k in range(0,24):
            hour_noise = np.random.normal(0,1,100)
            count = np.mean(hour_noise)
            while count < n_trans*day_distrib[k]:
                count = count+1
                indice_farmaco = random.randint(0,len(lista_farmaci)-1)
                record = np.array([df['PARTITAIVA'],lista_farmaci[indice_farmaco],random.randint(1,qta_max)])
                timestamp = giorno.replace(hour=k, minute=random.randint(0,59), second = random.randint(0,59))
                record = np.append(record,timestamp)
                lista.append(record)
    return lista
```
